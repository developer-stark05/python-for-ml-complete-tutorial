# Comparison operators:

The comparison operators in python are used to compare two values and after comparison, they return the result as a boolean.

There are 6 comparison operators in python:

|Operator | Name | Description |
|---|---|---|
| `<` | less than | used to test two values that whether the left number is lesser than the right. |
| `>` | greater than | used to test two values that whether the left value is greater than the left. |
| `<=` | less than or equal to | used to test two values that whether the left value is less than or equal to the right value. |
|`>=`|greater than or equal to| used to test two values that whether the left value is greater than or equal to the right value.|
| `==` | equal to | used to compare two values to confirm whether they are equal or not |
| `!=` | not equal to | used to compare two values whether they are not equal or not |

## <u> The '<' operator:</u>

The **<** is used to compare two values and confirms whether the left operand is less than the right. If the comparison is true, it returns 'True', otherwise it returns False. It returns False if the both operands are equal or the left operand is not greater than the right.

```python
print(4<5) # here 4 is less than 5 so the comparison is True.
number_1=5
number_2=4
print(number_1<number_2) # here 5 is greater than 4 so the comparison is False.
```

## <u> The > operator:</u>

The **>** operator is used to compare two values and confirms whether the left operand is greater than the right. If the comparison is true, it returns 'True', otherwise it returns False.

```python
print(4>2)  #here 4 is greater than 2 and the comparison is True.
left_number=6
right_number=7
print(left_number>right_number) #here 6 is greater than 7, so the comparison is False.
```

## <u> The <= operator </u>

The **<=** operator is used to compare two values and it confirms whether the left operand is less than or equal to the right by returning a boolean value.
If the comparison is true, it returns 'True', otherwise it returns False.

```python
print(4<=6) #here 4 is equal to 6 and the comparison is True
print(5<=5) #here 5 is not less than 5 but equal to 5. So, the comparison is True.
print(6<=3) #here 6 is neither less than or equal to 3, So, the comparison is False.
```

## <u>The >= operator </u>
The **>=** operator is used to compare two values and confirm that whether the left operator is greater than or equal to the right operator. It returns True if the comparison is True otherwise it returns False.

```python
print(5>=3) #here 5 is greater than 3, so it returns True.
print(3==3)  #here 3 is equal to 3. So, it returns True.
number_1=5
number_2=6
print(number_1>=number_2) #here 5 is not greater than or equal to 6. So it returns False.
```

## <u> The == operator </u>
The == operator is used to compare two values whether they are equal. And if they are equal, it returns True. Otherwise it returns False.

```python
print(4==4) #here 4 is equal to 4, so it returns True.
print(4.7==4.7) #here 4.7 is equal to 4.7, so it returns True.
```

## <u>The != operator </u>

The != operator is used to compare two values and confirm that whether they are not equal or not. If they are equal, they return True. Otherwise, they return False.

```python
print(5!=4) #Here 5 not equal to 4, so it returns True.
print(4!=4) #Here 4 is equal to 4, so it returns False.
```

## Precedency of comparison operators

The precedency of comparison operators goes like this:

|S.no|Operator|symbol|
|-----|-----|----|
|1|Less than operator|<|
|2|Greater than operator|>|
|3|Less than or equal to operator|<=|
|4|Greater than or equal to|>=|
|5|Equal to operator|==|
|6|Not equal to operator|!=|

## Comparison operators with strings:

The comparison operators are used with strings in which the comparison is done in lexicographical order. Lexicographical order means comparing strings character by character according to Unicode values.
Unicode is a large set of values where each individual symbol or a value has an integer symbol.
When we use comparison operators, the comparison is done character by character.
When we use < symbol for strings, the strings are compared character and if the character has the first position in alphabet than the other, then the first position character containing string is treated as the greater string and the comparison stops here because here the differnce occured between characters. If the comparison is correct, it returns True, otherwise it returns False.
Same happens with all other operators.

```python
print("apple"<"banana") #this returns True because here the comparison stops in the beginning itself because of characters difference and in alphabet, a has the place before b, so a is less than b so apple is less than banana.
print("bad">"aromatic") # comparison stops at the first character and b is greater than a so the comparison is True and it returns True.
print("")
```
