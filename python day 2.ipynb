# Python syntax and its characteristics
## Python syntax:
Syntax is a set of rules that define how the code should be written so it can be understood by the interpreter or compiler. Python syntax helps us to write a clean readable code in which it can be understood by the humans easily.
## Key characteristics of Python syntax:
* Case sensitivity: Python is a case sensitive in which case and Case are not same terms in python.
* Semicolons: Unlike other languages, we dont need to use semicolons at the end of the lines of the python code.
* Platform independent: Python is a platform independent language, meaning that the code can be run on any machine irrespective of the operating system    which has python installed in it. Languages like C, code is executed from compiler to cpu directly. So, for different machines and different operating   systems have different architectures. So the C code from one machine doesn't run on another due to different architectures. But python has an            advantage-BYTECODE. The bytecode is universal and same on every machine and also handled by interpreter so it cant be affected by the cpu                architectures. So write once-run anywhere.
* Code blocks: In most of the languages, the code blocks are represented by the flower brackets. But in python, we dont need to use brackets. We can seperate them by using indentation(4 spaces.)

```python
user=5
if user==7:
    print("hello")
```

here in the above program, the gap before the print statement is called indentation.
* Dynamic typing language: Unlike C and Java, as they are statically typed languages in which we should declare the datatype of the value we are going to store in the variable. But in python, as it is a dynamically typed language, we dont need to explicitly declare the datatype of the value which we are going to store in the variable.
