# Logical operators

Logical operators helps us to compare and evaluate multiple
conditions at once.

Just like comparison operators, logical operators also return boolean values True or False.

There are three main logical operators- not,and,or.

## The 'and' operator:

The and operator returns True only if both statements on either side are True. If even one side is false, the whole thing becomes false.

```python
x=10
print(x>5 and x<15)
print(x<10 and x>5)
```

## The or operator

The or operator returns True if atleast one comparison is True. It returns False only when the all conditions are False.

```python
x=10
print(x<3 or x>5)
print(x==6 or x==7)
```

## The not operator

The not operator simply reverses the logical state of the condition. If a result is True, not makes it False, and vice versa.

```python
print(not(True))
```

## Short-circuit evaluation

Python is efficient. When evaluating logical expressions, it often stops as soon as it knows the final answer.


1. For **and** operator: If the first comparison is False, Python doesnt look at the second expression. Because the whole thing must be false anyway.
2. For **or** operator: If the first comparison is True, Python doesnt look at the second expression. Because the whole thing must be True anyway.



## Operators precedency

The logical operators goes like this:

|logical operators|
:-----------------:
|not|
|and|
|or|

The not has the highest priority and then the and operator and then the or operator.

**Note**: Eventhough they follow the precedency, we can change the order of the evaluations by enclosing them in brackets.
