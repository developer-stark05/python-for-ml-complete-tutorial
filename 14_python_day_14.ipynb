## Iterative statements

Iterative statements, also called as loops, are the statements which helps us to execute a block of code multiple times. When we want to print a line 10 times, instead of writing it 10 times, we can use a loop to print it 10 times.

There are two types of loops in python:
* while loop
* for loop

### while loop

Syntax:

The while loop first checks the condition, then if its true, then it executes the block of code and again checks the condition. If its True, it again executes the block of code and the block of code execution continues until the condition becomes false.

But sometimes, if the condition doesnt become False forever, the loop keeps running forever. So to keep the loops not to become infinite loop forever, we need three parts:

**initialization:** The variable to start the count.

**Condition:** The condition to keep loop active.

**update:** changing the variable to make the condition false.

```python
variable=5
while variable>=0:
    print(variable)
    variable-=1
```

As if you see in the loop, the variable is initialized to 5 and the condition is variable not equal to 0. And if the condition is True, then the variable is printed and the variable value is decreased by 1 and then the variable value becomes 4 and then the condition is again checked. 4 is greater than 0. So again the code is executed and the value is decreased. And when the variable becomes less than 0, then the while loop stops.

We use while loop when we dont know how many times we need to run the loop.

```python
password=input("Enter your password:")
while password!='1234':
    password=input("Enter your password")
```

The above code asks the user for the password until the user enters the password as 1234.

# for loop

**for** loop is an iterator based loop. It loops through an iterator; it visits each character in the iterator and executes the block of code.

* **iterator** = collection of characters.

string, list, tuples,dictionary and sets are all iterators.

### syntax:

If we look at the syntax, variable is used because the for loop visits every character in the iterator and to apply the indented block of code, it should store the value in a variable. So variable is used and iterator is for the looping of for loop.

```python
sum=0
sequence=[1,2,3,4]
for number in sequence:
    sum+=number
print(sum)
```

If we see the execution of the above code, the sum variable is initialized to 0 and the variable number is for storing the iterating values and the sequence is the iterator and when the loop starts, then for loop visits every element and the element is stored in the number and then the number is added to the sum and in the next iteration, new number is stored and this process continues until it reaches the end of the iterator.

If we want to run the for loop a specific number of times, then we use range() function. range() function produces a sequence of numbers only when loop is running.

```python
for number in range(6):
    print(number,end=' ')
```

If we look at the code, here the range function containing the number 6, it means it contains the numbers from 0 to 5.
The range function handles the number production in the following way:
* range(stop) ------> contains the numbers from 0 to stop-1
* range(start, stop) ------> contains the numbers from the start number to the number stop-1 (stop number is not included in the sequence)
* range(start, stop, step) ------> contains the numbers from start to stop-1 and it skips the numbers by step.

```python
for number in range(6): # contains numbers from 0 to 5
    print(number, end=" ")
```

```python
for number in range(1,101): # contains numbers from 1 to 100
    print(number, end=' ')
```

```python
for number in range(0,11,2): # produces numbers from 0 to 10 by skipping 2 numbers
    print(number, end=' ')
```
