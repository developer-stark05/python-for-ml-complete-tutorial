## Arithmetic operators
Arithmetic operators are used to make arithmetic calculations with numbers. In python we can use arithmetic operators to make:

* Addition
* Subtraction
* Multiplication
* Divison
* Floor divison
* Modulus
* Exponentiation
These are called operators and the numbers we use with them in an operation are called operands.

| Operator  | Operation    | Explaination|
|:-----------|:------------:|------------:|
|    +       | Addition     | Adds the operands.      |
|    -    | Subtract         | Subtracts the operands.        |
|    *    | Multiplication   | Multiplies the operands. |
|    /    | Divison          | Divides the operands and returns the quotient in floating point number.    |
|   //    | Floor divison    | Divides the operands and returns the quotient in integer form. |
|   %     | Modulo operator  | Returns the remainder of the numbers when divided. |
|   **    | Exponentiation   | Calculates the value of a number when raised to a certain power |


## '+' operator application:
The sum operator helps us to add the numbers.

```python
a=5
b=2
print(f"{a+b=}")
print(1+2+3+4+4)
```

## '-' operator application:
The difference (or) subtract operator is used to subtract the numbers.

```python
a=5
b=2
print(f"{a-b=}")
print(1-2-3-4-5)
```

## '*' operator application:
The multiplication operator is used to multiply the numbers.

```python
a=5
b=2
print(f"{a*b=}")
print(1*2*3*4*5)
```

## '/' operator application:
The divison operator is used to divide the numbers.

```python
a=5
b=4
print(f"{a/b=}")
```

## '//' operator application:
The floor divison operator is used to divide the numbers and return the result as an integer even if the result is a floating point number.

```python
a=5
b=4
print(f"{a//b=}")
```

Observe the difference between the calculation. The same numbers are divided by divison operator and floor divison operator but got different results.
The floor divison operator rounds the result towards the negative infinity and then removes the decimal part and then returns the value.

## '%' operator application:
The modulo operator is used to return the remainder of the divison of the numbers. Here the modulo operator is returning remainder, so it always an integer.

```python
a=5
b=4
print(f"{a%b=}")
```

## '**' operator application:
The exponentiation operator is used to raise a number's power upto a number.

```python
print(2**3)
```

## Important points:
* The divison operator always return the answer in the form of floating point number even though the numbers are correctly divisible.

```python
print(4/2) # 4 is divided by 2 at 2. But the answer won't be 2. Its 2.0
```

* And when we add an integer and a floating point number, the answer will be always a floating point number. When python encounters the operands as an integer and a floating point number, the integer is converted into a floating point number temporarily and then they are added and the result will be a floating point number. This is called type promotion.

```python
print(2.35+2)
```

* Type promotion happens in other operations also, including subtraction, multiplication and divison and floor divison also. If any one of the operands
are of a floating point number, the rest numbers (if they are integers, converted into floating point numbers temporarily and the result will be a 
floating point number.)

```python
print(4-1.5)
print(1.9*4)
print(2.6/3)
print(4.5//3)
```

If we are asked to answer for a sum of 6+6 or 9+8+5 or difference of 9-5 or product of 6*7, we can answer them. But what if all they were mixed? 
I mean, what if you were asked to answer the result of 5*9+7-6?
Here the **precendency rules** come in. The precedency rules teaches us to evaluate the expression in a proper way when multiple operators are present.
| Precedence of Arithmetic Operators | Operator(s)                          | Associativity   |
|:-----------------------------------|:-------------------------------------|:----------------|
| P                                  | Parentheses                          | Left to Right   |
| E                                  | Exponentiation (**)                  | Right to Left   |
| M                                  | Multiplication (*)                   | Left to Right   |
| D                                  | Division (/), Floor Division (//), Modulo (%) | Left to Right   |
| A                                  | Addition (+)                         | Left to Right   |
| S                                  | Subtraction (-)                      | Left to Right   |

From the above table, we see that the parentheses have the highest precedency. That means, first the expression in the brackets should be executed. And if they are multiple brackets, start solving from the left to right. And after parentheses, exponents have the highest precedency. But for more exponents, execute from right to left. After exponentiation, multiplication and divison have same precedency. And similarly, if both are present in an expression, evaluate from left to right. After parentheses and exponentiation, multiplication and divison, lastly addition and subtraction have same precedency and evaluated from left to right.

```python
1-4*5+(4-7)+(4+6)
```

The above answer will be -12 because, according to our precedence rules, first parentheses should be executed. But we have two parentheses. But parentheses are executed from left to right. So, we execute the left parentheses first. 4-7=-3 and then 4+6=10. And now we have the multiplication as second highest precedency. So, 4*5=20. Now the expression is in the form of 1-20-3+10. Only addition and subtraction are present. So, execute them from left to right. 1-20=-19 and -19-3=-22 and -22+10=-12 which is our answer.
