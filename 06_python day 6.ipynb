## Assignment operators:

Python assignment operator allows us to assign a value to a variable, often combining an arithmetic or a bitwise operation.
The basic assignment operator is "=", which is used to assign a value to a variable.

## '=' operator :
The '=' is a basic assignment operator, which simply assigns a value to a variable.

```python
name="python" #"python" is assigned to the variable "name"
```

## '+=' operator:
The '+=' operator is an assignment operator in which it allows to add a value to the existing value in the variable.

```python
a=5
a+=6      #a+=6 equals to a=a+6 that means, now a= 5+6
print(a)
```

## '-=' operator:
The '-=' operator is an assignment operator in which allows to subtract a value to the existing value in the variable.

```python
a=5
a-=3      #a-=3 equals to a=a-3 that means, now a= 5-3
print(a)
```

## '*=' operator:
The '*=' operator is an assignment operator in which allows to multiply a value to the existing value in the variable.

```python
a=5
a*=3      #a*=3 equals to a=a*3 that means, now a= 5*3
print(a)
```

## '/=' operator:
The '/=' operator is an assignment operator in which allows to divide a value to the existing value in the variable.

```python
a=10
a/=5     #a/=5 equals to a=a/5 that means, now a=10/5
print(a)
```

## '//=' operator:
The '//=' operator is an assignment operator in which allows to perform floor divison with a value to the existing value in the variable.

```python
a=10
a//=5     #a//=5 equals to a=a//5 that means, now a=10//5
print(a)
```

## '%=' operator:

The '%=' operator is an assignment operator in which allows to perform modulo operation with a value to the existing value in the variable.



```python
a=10
a%=5      # a%=5 equals to a=a%5 that means, now a=10%5
print(a)
```

## '**=' operator:
The '**=' operator is an assignment operator in which allows to raise the power of present value in the variable to a certain number.

```python
a=10
a**=2    # a**=2 equals to a=a**2 that means, a=10**2
print(a)
```

**Notes:**

* Assignment operators and compound assignment operators are different. Assignment operator in python is only one. Thats equal to =. Its used to assign a value to a variable. But compound assignment operators are those who come with an arithmetic and assignment or bitwise or assignment operation. Those help us to perform operations on the values which are held by the variables.
* When we use a compound assignment operator, then the provided operation is done on the variable holding value. But the value in the variable doesnt change. Because, the integers are immutable. Actually when we use the compound assignment operators, the operation is done, and a new object is created and the result is stored in that new object. The creation and the operation are done simultaneously. And then the variable name is given to the new object and then the old object memory is cleaned up by the garbage memory collector.

* We cant use compound assignment operators without initalizing the variable. If we consider a variable x, and we use x+=2, python first considers the left operand and it finds x, which is a variable. And searches for the value in x, and here the x doesnt have any value yet. So python doesnt know how to handle this. So, python raises a NameError.

```python
x+=4
print(x)
```

* We cant chain compound assignment operators like a+=b+=4. Because chaining is only allowed for the operators which form expressions and expressions can return a value so they can be used further in program. But the compound assignment operators doesnt return values, they just update them. So, they cant be chained, and one more thing is a=b=c=3 works because its a statement in python syntax which allows us to assign a value to multiple variables at once.


* We cant write this a=(b=5). It results in a syntax error. Because, when python sees braces, it expects an expression is enclosed in it. Because braces are only dedicated for expressions. But there is no expression and its only the assignment in the bracket which is a statement and it results in a syntax error.
