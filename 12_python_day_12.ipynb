# Desicion making statements

The statements in the program which make desicions to execute a block of code according to the truth value of the logic assigned to it.

They are:
* If statements
* elif statements
* else statements
* nested if statements.

## if statements

The if statement is the statement which executes the block of code when the assigned condition is True and after the indented block of execution, it reaches the end of the statement.

```python
number=int(input("Enter a number:"))
if number%2==0:
  print(f"{number} is an even number.")
print("Finish")
```

If the condition is True, the indented block of the code is executed. Otherwise, the indented block is skipped and the end of the program is reached.

```python
number=int(input("Enter a number:"))
if number%2==0:
  print(f"{number} is an even number.")
print("Finish")
```

In the above if statement, when 2 is entered, due to its an even number, the indented block of code is executed. But when its 7, its odd number and the indented block is skipped.

## elif statements

If we need to check more than one condition, we need to use elif statements. Elif statements are executed when the if statement indented code block execution is skipped.

```python
number=int(input("Enter a number:"))
if number%2==0:
  print(f"{number} is an even number.")
elif number%3==0:
  print(f"{number} is divisible by 3.")
print("Finish")
```

In the above program when 3 is entered, first the if statement condition is checked. And 3 is an odd number, its not divisible by 2. And then due to the if statement execution is skipped, then the elif statement execution is done. And here in the elif statement, the condition is True, so its executed.

## else statement

When the if statement and all the elif statements are False, then the else block code is executed.

```python
number=int(input("Enter a number:"))
if number%2==0:
  print(f"{number} is an even number.")
elif number%3==0:
  print(f"{number} is divisible by 3.")
else:
  print("Finish")
```

Unlike if and elif statements, else statement doesnt have any condition. It just executes when all the statements are False.

In the above code, when 11 is entered, the if statement is checked first, 11 is not divisible by 2, then the elif condition is checked, 11 is not divisible by 3 too. The if block, elif block are skipped. So, the else block is executed. So it directly printed 'Finish'.
all the elif blocks and else blocks should start only after the if statement.
And only one if statement and one else statement should present in a desicion making statement program.
And else statement is not mandatory always. But if statement is always mandatory. And one more thing is we can write as many elif statements we want but the if statements and else statements should be only one.
In desicion making statements, if one of the statement is executed, then automatically the program is reached to the end of the program no matter even the other statements are true.

```python
number=int(input("Enter a number:"))
if number%2==0:
  print(f"{number} is an even number.")
elif number==0:
  print(f"{number} is a 0.")
print("Finish")
```

In the above code, when we enter 0, in the if statement, we have the condition that if that number is divisible by 2, print the number and say that its an even number. And 0 is an even number, so the print statement is executed. But if we look at the elif condition, its condition of number==0 is also true because, we entered 0. But already the if statement is True, so the rest statements execution is skipped.

## Nested if statements

Nested if statements are the statements which are used when we are checking multiple conditions and the user should satisfy multiple conditions. Nested if simply mean if inside if. We can write as many nested if statements we want.

```python
marks=int(input("Enter your marks:"))
if marks>90:
  age=int(input("Enter your age:"))
  if age>18:
    handicap_status=input("Are you handicaped(Yes/No):").title()
    if handicap_status=='No':
      print("You are elligible for sports.")
    else:
      print("Sorry, you are not eligible for sports due to your handicap status.")
  else:
    print("You are not elligible for sports due to your age.")
else:
  print("You are not elligible for sports due to your low marks.")
```

If you observe in the above program, to elligible to sports, the candidate must satisfy three conditions compulsory:
* He must have marks greater than 90.
* His age should be more than 18.
* He should not be handicaped.

If one of this condition fails, the student cant be elligible for the sports.
So in this situation, we use nested if statements.
Its not necessary to have else statement for every if statement in nested if statement. But else statement is highly recommended.
