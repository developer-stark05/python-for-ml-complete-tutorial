# <u> Membership operators </u>

The membership operators are used to check whether a value or a variable is present in the given sequence or not. The sequence is the collection of characters. 
> Examples for sequences are: strings, lists, tuples, dictionaries, sets.

There are two membership operators in Python.
* **in** operator.
* **not in** operator.

## The in operator:
* The **in** operator is used to check whether the given value is present in the given sequence or not. If the value is present, it will return **True**
  otherwise it returns **False**.

```python
name="Srivathsav"
print("vathsav" in name)
print("SNS" in name)
```

## The not in operator:
* The **not in** operator is used to check whether the given value is not present in the sequence. Its opposite of the in operator. The not in operator will return **True** if the value is not present in the given sequence otherwise it will return **False**.

```python
name="Srivathsav"
print("SNS" not in name)
print("vathsav" not in name)
```

## Note:
* The membership operators are case sensitive. So the values "Sri" and "SRI" and "sRI" and "sri" are different values.

```python
sequence="Mathematics"
print("Math" in sequence)
print("math" in sequence)
print("MaTH" in sequence)
```
