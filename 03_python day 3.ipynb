# Datatypes in Python
Python has several built in datatypes that helps you to store and manipulate different kinds of information. Lets explore each type of datatype.
## Numeric types:
Numeric types are used to store numbers.
* int type(integers):
  - Whole numbers without decimal part are considered as int type values.
  - Can be positive, negative or  zero.
  - Used for counting, indexing and whole number calculations.

* float type(floating point numbers):
  - Decimal numbers with decimal part are considered as float type values.
  - Can be positive, negative or zero.
  - Used for accurate decimal calculations and measurements.

* complex type(complex numbers):
  - Numbers with real and imaginary type
  - written in the form of a+jb
    - a is the real part
    - b is the imaginary part
    - j represents the imaginary part
  - Used in advanced mathematical calculations.

## Text types:
Text types are used for text storage and manipulation.
* str type(strings):
  - sequence of characters.
  - Any text or numbers or symbols which are enclosed in single quotation marks '' or double quotation marks "" are called strings.
  - Can be written upto multiple lines by using triple quotes

## Sequence types:
Sequence types in Python are ordered collections of items which can be accessed by their index value.
* Lists:
  - Ordered collections of items of diffrent types.
  - Written in square brackets [].
  - Used for collections that need to be modified.

* Tuples:
  - Ordered, immutable collections of items of different types.
  - Written in parentheses ().
  - Cannot be modified after creation.
  - Used for the collections that should not be modified.

## Mapping types:
Mapping types are the datatypes that store collections of key value pairs where each key maps to a specific value. Same as in dictionary, you search for a meaning with the help of the word name, in python, mapping type datatype stores a value by assigning it to a unique value called key.
* dict (Dictionaries):
  - ordered collection of key value pairs
  - written in curly brackets {}.
  - keys must be unique from all existing keys in the dictionary.
  - Values can be of any datatype.


## Set types:
Set types are used to store an unordered collection of unique elements.
* set(Sets):
  - Unordered collection of different datatype values.
  - written using set() or {}.
  - cannot contain duplicate values(no two values should be same).
  - Used for fast membership testing and unique item collections.

* frozenset(Frozenset):
  - Immutable sets
  - Written using frozenset()
  - Cannot be modified after creation
  - Used when you need an immutable collection of unique items.

## Boolean type:
A datatype in python that represents logical values that can only be True and False.
* bool(Boolean):
  - represents the logical values and those are "True" and "False".
  - Used for conditional statements and logical operations.
  - Can be used in mathematical operations(True=1, False=0).

## Nonetype:
Nonetype represents a special datatype in python which is none, which represents the absence of the value.
* None(None):
  - A special type with a single value None.
  - Represents the absence of a value.
  - Used as a placeholder to indicate the absence of a value.
  - Often used a default return values for functions.
