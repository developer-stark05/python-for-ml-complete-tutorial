Before we learn programming, we should know what happens to the code when it is entered in the computer.
## Hardware architecture:
In the whole computer, only three parts are used to handle our program code. They are:
* Secondary Storage: To execute a program, we must store it on our computer first. After we write the program on our computer, the program gets stored     in the secondary storage to use it in the future.
* Primary Memory(RAM): When we click the program to run it, the program comes out of the secondary storage and gets loaded into the primary memory         called RAM. The primary memory is used because the CPU has an easy access to the program when its loaded into the primary memory. 
* CPU(Central Processing Unit): After the program is loaded from the secondary storage to the RAM, then the code is accessed by the CPU, which is the brain of the coumputer, the cpu is the one who executes the program code.
## Execution of the code:
When you type your code and save it, then it will be saved in the secondary storage and then when you click it to run, then the program is loaded into the computer RAM, and then the CPU will access the program code from the RAM and then the code will be executed by the CPU and the output is showed on the screen.

According to the closeness to the hardware and human readibility, the programming languages are classified into three types of programming languages:
* Low-level programming language: Directly interaction with hardware, the code isnt readable for humans.
* Mid-level programming language: Balances the interaction with hardware, while readable to humans.
* High level programming language: Programming language code is close to the language which is spoken by humans and can be understood by humans          easily. The interaction with the hardware is highly abstracted.

## Compilation vs Interpretation:
Compilation and Interpretation are two different ways of program execution. We will see about both of them.
* Compilation: The program code is converted into the machine language code(binary) completely. After the conversion, the code is packed in a file and when we run that file, then the code will be executed.
* Interpretation: The program execution is done line by line, one line is converted into machine language code and then the machine language code is     executed and output is showed on the screen and then the second line execution starts. No need of packing the machine code into a seperate file.
## What's the difference???
Compilation is fast because all the program code is converted into machine code at once and when we run that binary file, the execution is done without any other conversion. But its less flexible, if an error occured in the program, it can only be found after compiling the whole file and we can't know the exact place where the error occured.
Interpretation is little bit slow because only one line is converted into machine language code and that machine code is executed and output is showed to the screen and then next line. Unlike compilation, no binary code file is needed. But its flexible because when the error is occured, if we have a 10 lines of code, if there is an error at 6th line, the execution stops at 5th line and then the rest of the code isn't executed.

## Python code execution:
Python is an interpreted language. We must know how the python code is executed so we can know how the interpreted language codes are executed:
* The first program line is first taken by the interpreter, then instead of converted into machine code, it is converted into a rough version called     bytecode.
* Then this bytecode is send to a special program called VM(Virtual Machine), and its python's VM, so its called PVM. This PVM converts the bytecode into the machine language code.
* Then the machine code is then executed by the CPU and the output is displayed and the next line is selected for the execution.
Not only python, every interpreted language code is executed like this.
