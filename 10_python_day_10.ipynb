#<u>Identity operators</u>

Identity operators are used to check whether the variables are pointing to the same object in the memory.

Before we get started, we must know how the values in memory are stored in python.

## <u>The memory allocation</u>

When we create the variables in the memory, the values are stored in computer's RAM.
So when we create a variable like:

`a=8`

The value 8 is stored in an object in the memory(RAM has the memory location boxes in which the values are stored. Those boxes are called objects.) Then the variable name 'a' is attached on the object. And if i write:

`a=8`

`b=a`

No new object is created for 'b'. Simply the existing object 8 which is having the name 'a' gets another name 'b'. Now we can say that the variable 'b' and 'a' are pointing to the same variable in the memory.
If i write:

`a=8`

`b=a`

`a+=2`

We already know that integers are immutable, so when we try to add  2 to the existing value present in a, then instead of modifying the value in a, python creates a new object in memory and sums the 8+2 simultaneously and then the 10 is stored in this new object and the variable name 'a' is taken from the 8 memory object and given to the 10 memory object but the 8 memory object is not deleted because still the b variable name is sticked to the 8 memory object.

There are two identity operators in python. They are:

* **is** operator.
* **is not** operator

## <u>is operator</u>

The **is** operator is used to check whether two variables are pointing to the same object in the memory or not. If yes, it returns True. Otherwise it returns False.

```python
a=23
b=a
print(a is b) # check whether a and b are pointing to the same object in memory.
```

The above code returns True because a is directly assigned to b. That means the object 23 has 2 labels now, a and b. So, a is b is True. But sometimes we need to check the variables differently, when the variables are not assigned to each other, like below.

```python
a=23
b=23
print(a is b)
```

Here a is not assigned to b. But still it returns a is b as True. Here the concept of **integer interning** comes.

## <u>Integer interning</u>:

Before we move to know about integer interning, first we know what is interning. Interning means using only one copy even though we assign the value to multiple variables.

**Integer interning** means that when python starts, it pre allocates objects for the numbers from -5 to 256 in the RAM. So when we assign a number from this range to any different variables, the predefined object gets the variable labels but the object is not defined again. As we seen above, 23 is in the range of -5 to 256. So the object 23 is predefined and the object just get the labels a and b. But if we cross the range, the seperate objects will be created and is operator doesnt return True.

```python
a=257 # out of the range of -5 to 256
b=257
print(a is b)
```

## <u>Floating point interning</u>:

In python, floating point numbers doesnt interned as integers. Because if we select a specific range of floating point numbers to preallocate, just in the range of 1 and 2, there are infinite possible floating point numbers, and actually the integers interning range is from -5 to 256 because these numbers are used frequently. But if we intern floating point numbers and select a range of 1 and 2, there will be an infinite possibilities and its difficult to say which number you use more. So floating point numbers are not interned. So even though you assign the same value for different variables, they create seperate objects in memory.

```python
a=67.6
b=67.6
print(a is b)
```

## <u>String interning</u>:

Strings are interned according to the following rules:
* It should not be created in runtime.
* It should contain only letters or numbers or underscores.
* It should be short or single character.
If a string accepts one of the last two conditions and the first conditions, then the string is interned.

```python
string="python" #short and contains letters only
word="python"
print(string is word)
```

* The strings are not interned if they contain special symbols.

```python
string="python+"
word="python+"
print(string is word)
```

## <u>id() function:</u>

When a object is created in the memory, every object is given an individual integer, which is its memory adress. That memory adress can be seen by using
id() function. When we are using two variables, say **a** and **b**, if we are using a is b, then we are saying id(a) is id(b).
If the a and b are referring to the same object in memory, then the both variable's id() will be same. If ids are same, then the variables are pointing to the same object in the memory.

```python
a=67
b=66+1
print(id(a),id(b)) # memory adresses of the objects of a and b
print(a is b) # True because memory adresses of the objects are same.
```

```python
a=4.5
b=4.5
print(id(a),id(b)) # memory adresses of the objects of a and b
print(a is b)
"""False because memory adresses of the objects are not same. Float objects are not interned."""
```
