## print() function:
The **print function** is the basic function in python in which it is used to display a text or any type of information to the screen. Whatever we write in the brackets of the print() function and run it, that information will be printed to the screen.
## Syntax:

> print("In between these brackets, we can write the information we want to print.")

```python
print(12345) #printing a number
print(4.35) #printing a floating number
print("hello") #printing a string
```

We can use **print()** function to print the sum or difference of the numbers.

```python
print(2+4)
print(6-5)
```

## Printing multiple terms:
We can print multiple terms using a single print function.

```python
print("Hello!","My","name","is","Srivathsav")
```

Using of comma after the string tesults a single space in output. We can use the print like above to print multiple terms.
## Some useful print() function shortcuts:

* Changing the seperator:

  If we want to print multiple terms in a single print() function, but we want to seperate them with some distinct symbols. So, we use **sep**         parameter.

```python
print("apple","banana","pineapple",sep="-")
```

As shown in the above, 'sep' parameter is used to seperate the terms with our desired symbols.

Remember a thing that each print() function starts printing in a new line. It means, if we write 5 print() functions, the information in them is printed in 5 different lines.

```python
print("Python is created by Guido Van Rossum.")
print("Python is a beginner friendly programming language.")
print("I am a self taught programmer.")
```

As you observed, the information in three print() are printed in three distinct lines. But what if we want all those three lines in one line??

* Changing the **end()** character:

  If we want to print the sentences from a number of print() statements on a same line, then we need to use **end** parameter.

```python
print("Python is created by Guido Van Rossum.",end=" ")
print("Python is a beginner friendly programming language.", end=" ")
print("I am a self taught programmer.")
```

Here as we see, before without using the **end** parameter, the lines are printed in distinct and seperate lines. But after using "end", they are printed on the same line. And not only for printing the lines on the same line, we can also use end for make the sentence to print with a distinct end character. The information we write in the quotes of the end parameter, will be used at the end of the statement in the print() function.

```python
print("I am alright",end="....")
```

Here i wrote "...." in the quotes of the **end** parameter, so that symbol is added at the end of the sentence which is written in the print statement.
Each **print()** function starts a newline unless you use **end** parameter.

## Input() function:

The **input()** function is used to take input from the user via keyboard. It pauses the program, displays an optional prompt message, waits for the user to enter a value and returns the value as a string by default.

## Syntax:

> input("prompt")

**Prompt**: Prompt is a message displayed on the screen before the user enters the screen.

```python
input("Enter your name:")
```

## How input() function works?

* Displays the prompt message(prompt message is optional, not compulsory.)
* Waits for the user to enter the input.
* After then, returns the entered value as a string by default.

We can use **input()** function and **print()** function both to print our desired value.

```python
print(input("Enter what you want to print:"))
```

## Small program using input, print, and variables

We are going to write our very first program with our existing knowledge which asks the user for his name and age and phone number and adress as input and prints it to the screen.

```python
name=input("Enter your name:")
age=input("Enter your age:")
phone_number=input("Enter your phone number:")
adress=input("Enter your adress:")
print("Your name is",name)
print("Your age is",age)
print("Your phone number is",phone_number)
print("Your adress is",adress)
```

As we see in the above program, we used the variables to store the values and input function to take the values and print the values and but if we look at the print statements, instead of writing the sentences and the variable names seperately, we can embed the variable names into the sentences by using **f-strings**.

## F-strings:
F strings are special type of strings in python, which are prefixed with f or F, which we can embed variables easily so we can add the values from the variables into the sentences.

Lets look at the syntax of the first program again so we can try the f-strings on them.

```python
name=input("Enter your name:")
age=input("Enter your age:")
phone_number=input("Enter your phone number:")
adress=input("Enter your adress:")
print(f"Your name is {name}.")
print(f"Your age is {age}.")
print(f"Your phone number is {phone_number}.")
print(f"Your adress is {adress}.")
```

Now if we look at the code again, the print statements look cleaner than before. We should prefix "f" or "F" to a string to make it a f-string. Then we use curly brackets to embed the variable name in it so, when printing the string, the value stored in the variable name is printed instead of the variable name. 
## Some f-string short-cuts:
In f-strings, we can able embed the sum and difference the numbers using f-strings.

```python
print(f"The sum of 5 and 4 is {5+4}.")
```

In f-strings, we can able to format the floating point numbers also.

```python
price=6.7889956
print(f"The price when formatted is {price:.2f}.")
```

## String methods:
If we want to convert a string into capital letters, we will use **.upper()** method.

```python
print("srivathsav".upper())
```

If we want to convert a string into lowercase letters, we will use **.lower()** method.

```python
print("SRIVATHSAV".lower())
```

If we want to convert into a sentence into title case, we use **.title()** method so each new word's first word will be capitalized like the title.

```python
print("i am kani srivathsav".title())
```

If we want to capitalize the sentence, we use **.capitalize()** method. So, the first letter of the sentence is capitalized.

```python
print("i am a self-taught programmer.".capitalize())
```

Sometimes, in user input, unwanted gaps may enter and it may cause issues. So, we use strip methods for strings. I mean, if the name should be entered like "name", we may enter "   name". So to handle this, we have strip methods.

To remove the unwanted gaps present on the left of the string, we use **.lstrip()** method.

```python
name=(input("Enter your name:")).lstrip()
print(name)
```

Similarly, to remove the unwanted gaps present on the right of the string, we use **.rstrip()** method.

```python
name=(input("Enter your name:")).rstrip()
print(name)
```

And instead of .lstrip() and .rstrip(), we can use .strip() which handles gaps on both sides.

```python
name=(input("Enter your name:")).strip()
print(name)
```

**BONUS TIP**:

We can use string methods in f-strings.

```python
print(f"My name is {"kani srivathsav".upper()}.")
```

```python
print(f"{"this document is in jupyter notebook.".capitalize()}")
```

```python
print(f"{'    this is the gap added sentence       '.strip()}")
```
