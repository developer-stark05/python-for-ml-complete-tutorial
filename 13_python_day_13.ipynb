## Some useful built-in functions

These are some built-in functions which are frequently used in python:

* **abs()** function
* **min()** function
* **max()** function
* **sum()** function

### abs() function

The abs() function is used to calculate the absolute value of a number; it simply converts the negative number into positive and returns it. But it means that it returns the distance between the number and 0 on the number line.

```python
print(abs(6))
print(abs(-4.5))
```

### max() and min() functions

The max() and min() functions take the numbers as inputs and returns the greatest number and the smallest number of the given numbers.

```python
print(max(1,2,3))
print(min(9,9.6))
```

These functions work for the lists and tuples too.

```python
print(max([1,8,5]))
print(min([2,6,0]))
print(max((2,6,4)))
print(min((4.6,-3)))
```

### sum() function

The sum() function adds the numbers in a list or a tuple.

```python
print(sum([3,5,7]))
print(sum((5,7,8)))
```
